# Data Cleaning and Wrangling

#### USD needs to be converted to GBP

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('shopping_trends.csv')

# Define the conversion rate from USD to GBP
usd_to_gbp_rate = 0.78  # Example: 1 USD = 0.78 GBP

# Convert the `Purchase Amount (USD)` column to GBP
data['Purchase Amount (GBP)'] = data['Purchase Amount (USD)'] * usd_to_gbp_rate

# Display the updated dataset with both USD and GBP columns
print(data[['Purchase Amount (USD)', 'Purchase Amount (GBP)']].head())

# Save the updated dataset
data.to_csv('shopping_trends_in_GBP.csv', index=False)
```

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('shopping_trends.csv')

# Define the conversion rate from USD to GBP
usd_to_gbp_rate = 0.78  # Example: 1 USD = 0.78 GBP

# Convert the `Purchase Amount (USD)` column to GBP and create a new column
data['Purchase Amount (GBP)'] = data['Purchase Amount (USD)'] * usd_to_gbp_rate

# Drop the old 'Purchase Amount (USD)' column
data = data.drop(columns=['Purchase Amount (USD)'])

# Rename the 'Purchase Amount (GBP)' column to the old column name if needed
data = data.rename(columns={'Purchase Amount (GBP)': 'Purchase Amount (GBP)'})

# Display the updated dataset
print(data.head())

# Save the updated dataset
data.to_csv('shopping_trends_with_gbp.csv', index=False)
```

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('shopping_trends.csv')

# Define the conversion rate from USD to GBP
usd_to_gbp_rate = 0.78  # Example: 1 USD = 0.78 GBP

# Convert the `Purchase Amount (USD)` column to GBP and create a new column
data['Purchase Amount (GBP)'] = data['Purchase Amount (USD)'] * usd_to_gbp_rate

# Drop the old 'Purchase Amount (USD)' column
data = data.drop(columns=['Purchase Amount (USD)'])

# Rename the 'Purchase Amount (GBP)' column to the original column name
data = data.rename(columns={'Purchase Amount (GBP)': 'Purchase Amount (USD)'})

# Add the pound sign (£) to the 'Purchase Amount (USD)' column
data['Purchase Amount (GBP)'] = data['Purchase Amount (USD)'].apply(lambda x: f"£{x:.2f}")

# Display the updated dataset
from IPython.display import display
display(data)

# Save the updated dataset
data.to_csv('shopping_trends_with_gbp.csv', index=False)
```

## Analyzing the spending habits across age groups



### I believe the higher spending among the 51+ and 31-50 age groups can be attributed to a few factors:

### 1. Higher Disposable Income:
People in the 31-50 and 51+ age brackets are generally more likely to have established careers, higher salaries, and financial stability.
This results in greater disposable income, allowing them to make more purchases and spend more.

### 2. Family and Household Spending:
 Many individuals in the 31-50 range may have families, and thus they could be spending on household goods, kids' needs, or even for experiences such as travel.
 The 51+ age group might also be spending more on healthcare, leisure activities, or larger purchases like home renovations or investments.

### 3. Maturity and Experience:
 Older age groups tend to have more experience with managing finances, which can lead to better budgeting for purchases.
 Additionally, they might be more willing to make larger, one-time purchases, especially when they are in their peak earning years or are nearing retirement.

### 4. Less Spending Among Younger Age Groups:
 The 19-30 group may have more limited financial resources, especially if they're in school, early career stages, or dealing with student loans. 
 They may spend less overall compared to the older groups, with a larger portion of their income going toward essentials or saving for the future.
 The 0-18 group is likely to have the lowest spending, as they generally rely on parents or guardians for financial support.

#### Conclusion:
#### Based on the data, it's clear that the older age groups tend to have more financial stability and larger spending power, resulting in them being the largest contributors to overall spending.

# Data Visuliations 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plotting a bar plot for the 'Category' column
category_counts = data['Category'].value_counts()
category_counts.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Item Categories Purchased')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

# Analysis of the Bar Plot: Item Categories Purchased


The bar plot provides a visual representation of the overall number of purchases in each item category 
(e.g., Clothing, Footwear, Accessories). Each bar represents one category, and the height of the bar corresponds 
to the frequency of purchases made in that category.

Key Insights:
- Clothing is the most purchased category, as indicated by its significantly taller bar compared to the others.
- Accessories are the second most popular category, followed by Footwear and Outerwear, which have lower purchase counts.
- The differences in bar heights highlight the varying levels of customer interest or demand across these categories.

Business Implications:
- The high sales in the Clothing category suggest a strong customer preference, making it a key area to focus on for promotions or inventory expansion.
- Lower sales in Outerwear and Footwear might indicate opportunities to improve marketing strategies, introduce new products, or adjust pricing to boost interest.
- Understanding these purchase patterns enables better resource allocation and targeted marketing to maximize sales.



```python
# Pivoting the data for stacked bar plot
gender_category = data.groupby(['Gender', 'Category']).size().unstack()

# Plot stacked bar plot
gender_category.plot(kind='bar', stacked=True, colormap='Set2')
plt.title('Gender Distribution by Category')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

# Analysis of the Stacked Bar Plot: Gender Distribution by Category

The stacked bar plot illustrates the distribution of product category purchases by gender. Each bar represents one 
gender (Male or Female), and the segments within the bars correspond to the frequency of purchases in specific 
categories, such as Clothing, Footwear, Accessories, and Outerwear.

Key Insights:
- Males make a significantly higher number of purchases overall compared to females.
- The largest segment for males is Clothing, indicating a strong preference or demand for this category among male customers.
- Females, while contributing less overall, show a more balanced distribution across categories like Accessories and Footwear.

Business Implications:
- Businesses can focus promotions on clothing items targeted at male customers.
- Strategies can be developed to boost female engagement in categories where their participation is relatively lower.
- Insights into product preferences by gender can inform inventory planning and targeted advertising efforts.

# Box Plot Analysis: Purchase Amount (GBP) by Age Group

The box plot reveals consistent spending patterns across all age groups, with similar median purchase amounts and interquartile ranges (IQRs), indicating uniform behavior in the middle 50% of purchase amounts. The whiskers show comparable ranges of spending across groups, with outliers present in each, representing unusually high or low purchases. 

- **Younger individuals (18-25)** display a narrower range of spending.  
- **Middle-aged groups (46-55 and 56-65)** show slightly higher variability, possibly due to higher disposable incomes.  
- **The oldest group (65+)** exhibits stable spending patterns with fewer extreme outliers.  

These insights suggest that businesses can adopt uniform pricing strategies across demographics while leveraging high-value outliers through loyalty programs or premium offerings. Targeted campaigns for younger customers and middle-aged groups may also boost engagement and revenue.
